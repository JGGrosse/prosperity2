```python
#CLEANUP THE TRADE DATA

import pandas as pd
import numpy as np

round = '4'

trade_data = pd.read_csv(r'../trades_round_' + round + '.csv', delimiter=';').fillna(0)



if round == '1':
    trade_data_amethyst = pd.DataFrame(columns=trade_data.columns)
    trade_data_starfruit = pd.DataFrame(columns=trade_data.columns)

    for idx, row in trade_data.iterrows():              
        if row['symbol'] == 'AMETHYSTS':
            trade_data_amethyst.loc[idx] = row
        elif row['symbol'] == 'STARFRUIT':
            trade_data_starfruit.loc[idx] = row
    
    trade_data_amethyst = trade_data_amethyst.reset_index()
    trade_data_starfruit = trade_data_starfruit.reset_index()
    
    trade_data_amethyst = trade_data_amethyst.drop(['symbol', 'index'], axis=1)
    trade_data_starfruit = trade_data_starfruit.drop(['symbol', 'index'], axis=1)
    
    trade_data_amethyst['timestamp'] = ((trade_data_amethyst['timestamp'] / 100) + 1) + ((trade_data_amethyst['day'] + 2) * 10000)
    trade_data_starfruit['timestamp'] = ((trade_data_starfruit['timestamp'] / 100) + 1) + ((trade_data_starfruit['day'] + 2) * 10000)
    
    trade_data_amethyst.to_csv('amethyst_trade_data_all.csv', sep=';', header=trade_data_amethyst.columns, index=False)
    trade_data_starfruit.to_csv('starfruit_trade_data_all.csv', sep=';', header=trade_data_starfruit.columns, index=False)


elif round == '3':
    trade_data_chocolate = pd.DataFrame(columns=trade_data.columns)
    trade_data_strawberry = pd.DataFrame(columns=trade_data.columns)
    trade_data_roses = pd.DataFrame(columns=trade_data.columns)
    trade_data_basket = pd.DataFrame(columns=trade_data.columns)

    for idx, row in trade_data.iterrows():
        if row['symbol'] == 'CHOCOLATE':
            trade_data_chocolate.loc[idx] = row
        elif row['symbol'] == 'STRAWBERRIES':
            trade_data_strawberry.loc[idx] = row
        elif row['symbol'] == 'ROSES':
            trade_data_roses.loc[idx] = row
        elif row['symbol'] == 'GIFT_BASKET':
            trade_data_basket.loc[idx] = row
    
    trade_data_chocolate = trade_data_chocolate.reset_index()
    trade_data_strawberry = trade_data_strawberry.reset_index()
    trade_data_roses = trade_data_roses.reset_index()
    trade_data_basket = trade_data_basket.reset_index()
    
    trade_data_chocolate =trade_data_chocolate.drop(['symbol', 'index'], axis=1)
    trade_data_strawberry =trade_data_strawberry.drop(['symbol', 'index'], axis=1)
    trade_data_roses =trade_data_roses.drop(['symbol', 'index'], axis=1)
    trade_data_basket =trade_data_basket.drop(['symbol', 'index'], axis=1)
    
    trade_data_chocolate['timestamp'] = ((trade_data_chocolate['timestamp'] / 100) + 1) + ((trade_data_chocolate['day'] + 2) * 10000)
    trade_data_strawberry['timestamp'] = ((trade_data_strawberry['timestamp'] / 100) + 1) + ((trade_data_strawberry['day'] + 2) * 10000)
    trade_data_roses['timestamp'] = ((trade_data_roses['timestamp'] / 100) + 1) + ((trade_data_roses['day'] + 2) * 10000)
    trade_data_basket['timestamp'] = ((trade_data_basket['timestamp'] / 100) + 1) + ((trade_data_basket['day'] + 2) * 10000)
    
    trade_data_chocolate.to_csv('chocolate_trade_data_all.csv', sep=';', header=trade_data_chocolate.columns, index=False)
    trade_data_strawberry.to_csv('strawberry_trade_data_all.csv', sep=';', header=trade_data_strawberry.columns, index=False)
    trade_data_roses.to_csv('roses_trade_data_all.csv', sep=';', header=trade_data_roses.columns, index=False)
    trade_data_basket.to_csv('basket_trade_data_all.csv', sep=';', header=trade_data_basket.columns, index=False)
    
elif round == '4':

    trade_data_coconut = pd.DataFrame(columns=trade_data.columns)
    trade_data_coupon = pd.DataFrame(columns=trade_data.columns)

    for idx, row in trade_data.iterrows():
        if row['symbol'] == 'COCONUT':
            trade_data_coconut.loc[idx] = row
        elif row['symbol'] == 'COCONUT_COUPON':
            trade_data_coupon.loc[idx] = row
    
    trade_data_coconut = trade_data_coconut.reset_index()
    trade_data_coupon = trade_data_coupon.reset_index()

    trade_data_coconut = trade_data_coconut.drop(['symbol', 'index'], axis=1)
    trade_data_coupon = trade_data_coupon.drop(['symbol', 'index'], axis=1)


    trade_data_coconut['timestamp'] = ((trade_data_coconut['timestamp'] / 100) + 1) + ((trade_data_coconut['day'] + 2) * 10000)
    trade_data_coupon['timestamp'] = ((trade_data_coupon['timestamp'] / 100) + 1) + ((trade_data_coupon['day'] + 2) * 10000)

    trade_data_coconut.to_csv('coconut_trade_data_all.csv', sep=';', header=trade_data_coconut.columns, index=False)
    trade_data_coupon.to_csv('coupon_trade_data_all.csv', sep=';', header=trade_data_coupon.columns, index=False)


```
