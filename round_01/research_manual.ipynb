```python
import numpy as np
import pandas as pd

def scuba_pdf(X):
    return ((1/5000) * X) - 0.18

def scuba_cdf (X):
    return ((1/10000) * (X)**2) - 0.18 * X + 81


bid_1_lp = np.linspace(900, 999, 100)

max_profit_df = pd.DataFrame(columns=['bid_1', 'bid_2', 'max_profit'])

idx = 0

for bid_1 in bid_1_lp:
    
    bid_2_lp = np.linspace((int(bid_1) + 1), 1000, (1000 - int(bid_1)))
    
    for bid_2 in bid_2_lp:
        
        max_profit = ((1000 - bid_1) * scuba_cdf(bid_1)) + ((1000 - bid_2) * (scuba_cdf(bid_2)- scuba_cdf(bid_1)))
        
        max_profit_df.loc[idx] = [bid_1] + [bid_2] + [max_profit]
        
        idx += 1
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import matplotlib.pyplot as plt


X = list(max_profit_df['bid_1'])
Y = list(max_profit_df['bid_2'])
Z = list(max_profit_df['max_profit'])


fig = plt.figure(figsize=(20,12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(X, Y, Z, cmap='viridis', edgecolor='none')

max_profit_idx = max_profit_df['max_profit'].idxmax()

best_bid_1 = max_profit_df['bid_1'].iloc[max_profit_idx]
best_bid_2 = max_profit_df['bid_2'].iloc[max_profit_idx]
max_profit = max_profit_df['max_profit'].iloc[max_profit_idx]

ax.scatter(best_bid_1, best_bid_2, max_profit, color='red', s=50)

print(best_bid_1, best_bid_2, max_profit)

plt.show()
```
